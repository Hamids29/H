```python
import pandas as pd
import numpy as np
from collections import Counter 
import re
```

```python
# if we want to read csv files (usually larger)
#sharedwithmanager is from 2/25/24 - 3/6

#file_path = "sharedwithmanager.csv"  # Replace with your CSV file path
file_path = "10-7-6-6.csv"  # Replace with your CSV file path
test_df = pd.read_csv(file_path)
df = pd.read_csv(file_path)

call_trigger_df = test_df[test_df['Product'] == 'Call Trigger']

total_calls = test_df['Call: Call Name'].nunique()  # Count unique call names

# Count calls where Product is 'Call Trigger' and Key Message is 'PAS'
proactive_calls = len(
    test_df[
        (test_df['Product'] == 'Call Trigger') &
        (test_df['Key Message'] == 'PAS')
    ].drop_duplicates(subset=['Call: Call Name'])
)

proactive_calls_df = test_df[
        (test_df['Product'] == 'Call Trigger') &
        (test_df['Key Message'] == 'PAS')
    ].drop_duplicates(subset=['Call: Call Name'])


proactive_calls_id = proactive_calls_df['Call: Call Name'].unique()

calls_df = df[df['Call: Call Name'].isin(proactive_calls_id)]

#process and counts all SWM
def process_key_message(message):
    if isinstance(message, str):
        if re.search(r'(Proactive: SWM|Reactive: SWM).*', message):
            return 'SWM'
    return message

calls_df['Processed_Key_Message'] = calls_df['Key Message'].apply(process_key_message)

message_counts = calls_df['Processed_Key_Message'].value_counts()
unique_messages_per_call = calls_df.groupby(['Call: Call Name', 'Processed_Key_Message']).size().reset_index()
unique_messages_per_call.columns = ['Call: Call Name', 'Processed_Key_Message', 'Count']
# Count occurrences of processed messages related to proactive calls
message_counts = unique_messages_per_call['Processed_Key_Message'].value_counts()
top_messages = message_counts.head(3)
# Print results
print(f"Total number of calls: {total_calls}")
print(f"Number of Proactive Calls: {proactive_calls}")

print("\nTop 3 recurring key messages:")
for message, count in top_messages.items():
    print(f"- {message}: {count} occurrences")
```

```python
import pandas as pd

# Replace with your CSV file path
file_path = "1-1-6-6.csv"
df = pd.read_csv(file_path)

# Filter for 'Call Trigger'
call_trigger_df = df[df['Product'] == 'Call Trigger']

# Count unique call names
total_calls = df['Call: Call Name'].nunique()

# Count proactive calls where Product is 'Call Trigger' and Key Message is 'PAS'
proactive_calls_df = df[
    (df['Product'] == 'Call Trigger') &
    (df['Key Message'] == 'PAS')
].drop_duplicates(subset=['Call: Call Name'])

proactive_calls = len(proactive_calls_df)
proactive_call_ids = proactive_calls_df['Call: Call Name'].unique()

# Filter calls based on proactive call IDs
proactive_calls_data = df[df['Call: Call Name'].isin(proactive_call_ids)]

# Fixed process_key_message function with proper affordability check
def process_key_message(message):
    if isinstance(message, str):
        if "Proactive: SWM Affordability" in message or "Reactive: SWM Affordability" in message:
            return 'savingsprogram'
        elif "Proactive: SWM" in message or "Reactive: SWM" in message:
            return 'SWM'
        elif "Proactive: Monotherapy" in message or "Reactive Monotherapy" in message:
            return 'Monotherapy'
    return message

# Apply the key message processing to the entire dataframe
df['Processed_Key_Message'] = df['Key Message'].apply(process_key_message)

# Count unique calls with SWM Affordability messages
# This counts a call only once even if it has both Proactive and Reactive SWM Affordability
affordability_calls = df[
    (df['Key Message'].str.contains('Proactive: SWM Affordability', na=False) | 
     df['Key Message'].str.contains('Reactive: SWM Affordability', na=False))
]['Call: Call Name'].nunique()

# Count proactive calls with affordability messages
proactive_affordability_calls = proactive_calls_data[
    (proactive_calls_data['Key Message'].str.contains('Proactive: SWM Affordability', na=False) | 
     proactive_calls_data['Key Message'].str.contains('Reactive: SWM Affordability', na=False))
]['Call: Call Name'].nunique()

# Count all monotherapy calls
all_monotherapy_calls = df[df['Processed_Key_Message'] == 'Monotherapy']['Call: Call Name'].unique()
monotherapy_count = len(all_monotherapy_calls)

# Create a detailed breakdown of affordability messages
# This will show us how many calls have both proactive and reactive affordability messages
proactive_affordability = df[df['Key Message'].str.contains('Proactive: SWM Affordability', na=False)]['Call: Call Name'].unique()
reactive_affordability = df[df['Key Message'].str.contains('Reactive: SWM Affordability', na=False)]['Call: Call Name'].unique()

# Calls with both types
both_affordability = set(proactive_affordability).intersection(set(reactive_affordability))

# Print results
print(f"Total number of calls: {total_calls}")
print(f"Number of Proactive Calls: {proactive_calls}")
print(f"\nNumber of calls with Monotherapy messages: {monotherapy_count}")
print(f"\nNumber of calls with SWM Affordability (total): {affordability_calls}")
print(f"- Calls with Proactive SWM Affordability: {len(proactive_affordability)}")
print(f"- Calls with Reactive SWM Affordability: {len(reactive_affordability)}")
print(f"- Calls with both Proactive and Reactive SWM Affordability: {len(both_affordability)}")
print(f"Number of proactive calls with affordability messages: {proactive_affordability_calls}")
```
